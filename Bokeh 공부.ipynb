```python
import numpy as np
from bokeh.io import output_notebook, show
from bokeh.plotting import figure

output_notebook()
```

```python
# 스캐터 relation
p = figure(plot_width=400, plot_height=400)
```

```python
show(p)
```

```python
x=[0,1,2,3,4,5]
y=[0,1,2,3,4,5]
p.circle(x, y, size=10)
show(p)
```

```python
p.circle(x, y, size=10, line_color='orange', fill_color='navy')
show(p)
```

```python
p.circle(x, y, size=10, line_color='orange', fill_color='brown')
show(p)
```

```python
p.circle(x, y, size=10, line_color='orange', fill_color='blue', fill_alpha=0.5)
show(p)
```

```python
import pandas as pd
from bokeh.plotting import figure
from bokeh.sampledata.stocks import AAPL

df=pd.DataFrame(AAPL)
df
```

```python
df.index=pd.to_datetime(df['date'])
df=df.drop('date', axis=1)
```

```python
df
```

```python
p=figure(plot_width=500, plot_height=400, x_axis_type='datetime')
p.line(df.index, df['close'],color='navy', alpha=0.5)
show(p)
```

```python
p=figure(plot_width=700, plot_height=300, x_axis_type='datetime')
p.line([df.index, df['open']], df['close'], color='navy', alpha=0.5)
show(p)
```
